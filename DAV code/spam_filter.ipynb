```python
import numpy as np
import pandas as pd
import re
from collections import Counter

class SpamFilter:
    """
    A Naive Bayes classifier for detecting Spam vs Ham.
    """
    def __init__(self):
        self.spam_words = Counter()
        self.ham_words = Counter()
        self.total_spam = 0
        self.total_ham = 0

    def clean_text(self, text):
        text = str(text).lower()
        text = re.sub(r'[^a-z\s]', '', text)
        return text.split()

    def fit(self, texts, labels):
        """labels: 1 for spam, 0 for ham"""
        for text, label in zip(texts, labels):
            words = self.clean_text(text)
            if label == 1:
                self.spam_words.update(words)
                self.total_spam += 1
            else:
                self.ham_words.update(words)
                self.total_ham += 1
        print(f"Model trained on {len(labels)} samples.")

    def predict(self, text):
        words = self.clean_text(text)
        spam_score = 0
        ham_score = 0
        
        # Total word counts for denominator
        total_s_words = sum(self.spam_words.values())
        total_h_words = sum(self.ham_words.values())
        
        for word in words:
            # Laplace smoothing to avoid log(0)
            spam_score += np.log((self.spam_words[word] + 1) / (total_s_words + 1))
            ham_score += np.log((self.ham_words[word] + 1) / (total_h_words + 1))
            
        return 1 if spam_score > ham_score else 0

# --- CONFIGURATION ---
CONFIG = {
    "file_path": "emails.csv", 
    "text_col": "text",
    "label_col": "spam"
}

if __name__ == "__main__":
    # Example Usage
    model = SpamFilter()
    train_msgs = ["Get free coins now", "Hey are we meeting?", "Claim your prize", "Lunch at 12?"]
    train_labels = [1, 0, 1, 0]
    
    model.fit(train_msgs, train_labels)
    test = "Win a free gift card"
    result = "Spam" if model.predict(test) else "Ham"
    print(f"Result for '{test}': {result}")
```
